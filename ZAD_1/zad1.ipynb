```python
import cv2 as cv
import numpy as np
cap = cv.VideoCapture("movingball.mp4")

while cap.isOpened():
    ret , frame = cap.read()
    if not ret:
        break
    
    
    frame = cv.resize(frame,(480,640))
    
    hsv= cv.cvtColor(frame,cv.COLOR_BGR2HSV)
    
    lower_red_1 = np.array([0,90,50])
    lower_red_2 = np.array([10,255,255])
    upper_red_1 = np.array([170,90,50])
    upper_red_2 = np.array([180,255,255])
    
    #mask
    
    mask_1 = cv.inRange(hsv,lower_red_1,lower_red_2)
    mask_2 = cv.inRange(hsv,upper_red_1,upper_red_2)
    mask = mask_1|mask_2
    
    kernel = np.ones((7,7),np.uint8)
  #  mask = cv.morphologyEx(mask,cv.MORPH_OPEN,kernel)
    mask = cv.morphologyEx(mask,cv.MORPH_CLOSE,kernel)

    erosion = cv.erode(mask,kernel,iterations=1)
    dilation = cv.dilate(erosion,kernel,iterations=1)
    
    res = cv.bitwise_and(frame,frame,mask=dilation)
    
    M = cv.moments(dilation)
    
    cX = int(M["m10"]/M["m00"])
    cY = int(M["m01"]/M["m00"])
    cv.line(frame,(cX,cY-5),(cX,cY+5),(0,255,0),2)
    cv.line(frame,(cX-5,cY),(cX+5,cY),(0,255,0),2)
    
    cv.line(res,(cX,cY-5),(cX,cY+5),(0,255,0),2)
    cv.line(res,(cX-5,cY),(cX+5,cY),(0,255,0),2)
   # cv.circle(frame,(cX,cY),5,(0,255,0),-1)
    
    cv.imshow('res',res)
    cv.imshow('frame',frame)
    
    
    if cv.waitKey(1) == ord('q'):
        break
    
cap.release()
cv.destroyAllWindows()

        
```
